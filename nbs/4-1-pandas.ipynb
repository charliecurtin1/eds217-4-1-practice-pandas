## Practice 5

```python
# A - Add a column 'NET_SW' to bsrn with the net shortwave radiation
bsrn['NET_SW'] = bsrn['SWD_Wm2'] - bsrn['SWU_Wm2']

bsrn.head()
```

```python
# B - add a column 'NET_LW' to bsrn with the net longwave radiation
bsrn['NET_LW'] = bsrn['LWD_Wm2'] - bsrn['LWU_Wm2']
```

```python
# C - add a column 'NET_RAD' to bsrn with the net total radiation
bsrn['NET_RAD'] = bsrn['NET_SW'] + bsrn['NET_LW']

bsrn.info()
```

```python
# D - create a new dataframe with the day of the month and the daily mean values of shortwave incoming and outgoing, and longwave incoming and outgoing, and net total radiation

# To get you started, here's some example code to create a new dataframe containing each day, and the average incoming shortwave (solar) radiation, avg_SWD (aka SW_in),  and the average outgoing shortwave (solar) radiation, avg_SWU (aka SW_out):

daily_rad = []
```

```python
# using masking 
daily_rad = []

for d in bsrn.index.day.unique():
    avg_LWD = bsrn.LWD_Wm2[bsrn.index.day == d].mean()
    avg_LWU = bsrn.LWU_Wm2[bsrn.index.day == d].mean()
    avg_SWD = bsrn.SWD_Wm2[bsrn.index.day == d].mean()
    avg_SWU = bsrn.SWU_Wm2[bsrn.index.day == d].mean()
    avg_NET = bsrn.NET_RAD[bsrn.index.day == d].mean()
    # Append the current day two values to our list of daily radiation values:
    daily_rad.append([d, avg_LWD, avg_LWU, avg_SWD, avg_SWU, avg_NET])

daily_rad = pd.DataFrame(daily_rad, columns=['day', 'LW_in', 'LW_out', 'SW_in', 'SW_out', 'Net'])

daily_rad
```

## Practice 6

```python
# A - use a list of lists to construct a dataframe from the data in the table (in the class practice)
df1 = pd.DataFrame([['Amazon', 6400, 7050000],
                   ['Congo', 4371, 4014500],
                   ['Yangtze', 6418, 1808500],
                   ['Mississippi', 3730, 3202230]],
                   columns=['River', 'Length(km)', 'Drainage area(km^2)'])

df1
```

```python
# B - use a dict to construct a dataframe df2 containing the data in the table below 
df2 = {
    'River': ['Zambezi', 'Mekong', 'Murray', 'Rhone', 'Cubango'],
    'Length(km)': [2574, 4023, 2508, 813, 1056],
    'Drainage area(km^2)': [1331000, 811000, 1061469, 98000, 530000]
}
df2 = pd.DataFrame(df2)
```

```python
# C - concatenate df1 and df2 into a new dataframe with all 9 rivers

df3 = pd.concat([df1, df2])

df3.reset_index(inplace = True, drop = True)

df3
```

```python
# D - create a new dataframe "rivers" with the discharge, mouth, source, and continent information and add this to the dataframe from A with all of the data in the table below

rivers = pd.DataFrame({'Discharge (m^2/s)':[209000,41200,30166,16792,3400,16000,767,1710,475],
              'Mouth':['Atlantic Ocean','Atlantic Ocean','East China Sea','Gulf of Mexico','Indian Ocean','South China Sea','Southern Ocean','Mediterranean Sea','Okavango Delta'],
              'Source':['Rio Mantaro', 'Lualaba River','Jianggendiru Glacier','Lake Itasca','Miombo Woodlands','Lasagongma Spring','Australian Alps','Rhne Glacier','Bi Plateau'],
              'Continent':['South America', 'Africa','Asia','North America','Africa','Asia','Oceania','Europe','Africa']})

rivers
```

```python
# D - join the rivers dataframe with df3 
# axis = 1 argument tells concat to append the values as columns 
df4 = pd.concat([df3, rivers], axis = 1)

df4
```

## Practice 7

```python
# Create a plot with the rivers data
import matplotlib.pyplot as plt
```

```python
# creates a figure and axis
fig, ax = plt.subplots(figsize = (10, 6))

# adjust bar width and bar_colors to be river-themed
bar_colors = ['#0066cc', '#0077b3', '#0099cc', '#00aadd'] 
bar_width = 0.6

# customize the plot and axes
ax.bar(df4['River'], df4['Drainage area(km^2)'], 
       color = bar_colors, width = bar_width, 
       edgecolor = 'black')

# add labels and title
ax.set_xlabel('River', fontsize = 14)
ax.set_ylabel('Drainage Area (km^2)', fontsize = 14)
ax.set_title('River Drainage Areas', fontsize = 16)

# set grid lines behind the bars
ax.set_axisbelow(True)

# customize grid lines
ax.grid(axis='y', linestyle='--', 
        alpha=0.6)

# show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# stylized using chatgpt
fig, ax = plt.subplots()

bar_width = 0.5

for i, (river, area) in enumerate(zip(df4['River'], df4['Drainage area(km^2)'])):
    x = [i - bar_width / 2, i + bar_width / 2]
    y = [0, area]
    flare_width = area * 0.002  # Adjust this factor to control the flare width
    flare_x = np.linspace(x[0], x[1], 100)
    flare_y = area * (1 - (flare_x - x[0]) ** 2 / (x[1] - x[0]) ** 2)  # Flared ends
    ax.fill_between(flare_x, flare_y, color='blue', alpha=0.7)

ax.set_xlabel('River')
ax.set_ylabel('Drainage Area (km^2)')
ax.set_title('River Drainage Areas')

ax.set_xticks(np.arange(len(df4['River'])))
ax.set_xticklabels(df4['River'])

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
